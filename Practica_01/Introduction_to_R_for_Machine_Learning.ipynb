###  Create an array of 10 random integers between 1 and 100, and then calculate the mean and standard deviation of the array.

```r
set.seed(0)

array <- sample(1:100, 10, replace = TRUE)

mean_value <- mean(array)
sd_value <- sd(array)

print(array)
print(paste("Mean:", mean_value))
print(paste("Standard Deviation:", sd_value))
```

### Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column.

```r
set.seed(0)  

array <- matrix(sample(1:100, 3*4, replace = TRUE), nrow = 3, ncol = 4)

row_sums <- rowSums(array)
col_sums <- colSums(array)

print(array)
print("Sum of each row:")
print(row_sums)
print("Sum of each column:")
print(col_sums)
```

### Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than or equal to 30

```r
df <- data.frame(
  Name = c("Alejanro", "Isaac", "Belen", "Gutenberg", "Ivan"),
  Age = c(25, 32, 45, 28, 38),
  City = c("Quito", "Guayakill", "Manta", "Quito", "Machachi")
)

filtered_df <- subset(df, Age >= 30)

print(filtered_df)
```

### Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame.

```r
df <- read.csv("/kaggle/input/riyadh-cafes/riyadh_cafes.csv")

# Calculate the mean, median, and mode of one of the columns
column_name <- "rating_count"  # Replace 'your_column' with the actual column name in your DataFrame

mean_value <- mean(df[[column_name]])
median_value <- median(df[[column_name]])
mode_value <- names(which.max(table(df[[column_name]])))


print(paste("Mean:", mean_value))
print(paste("Median:", median_value))
print(paste("Mode:", mode_value))
```

### Create a scatter plot of random x and y values between 1 and 100.

```r
set.seed(0)  

x <- runif(50, min = 1, max = 100)  
y <- runif(50, min = 1, max = 100)  

# Create the scatter plot
plot(x, y, main = "Random Scatter Plot", xlab = "X", ylab = "Y")
```
